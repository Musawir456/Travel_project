# 4) Visualization (Graphs)

```python
def plot_avg_popularity_by_destination(avg_dest_df, top_n=10):
    """
    Top N destinations by average popularity ke liye bar chart.
    """
    data = avg_dest_df.head(top_n)
    plt.figure(figsize=(10, 6))
    sns.barplot(x="Name", y="Popularity", data=data, palette="viridis")
    plt.xticks(rotation=45, ha="right")
    plt.title(f"Top {top_n} Destinations by Average Popularity")
    plt.tight_layout()
    plt.show()
```

# 5) Simple recommendation function

```python
def recommend_destination(df, best_time=None, dest_type=None, top_n=5):
    filtered = df.copy()

    if best_time:
        filtered = filtered[filtered["BestTimeToVisit"] == best_time]

    if dest_type:
        filtered = filtered[filtered["Type"] == dest_type]

    if filtered.empty:
        print("Koi destination nahi mila given filters ke saath.\n")
        return filtered

    result = filtered.sort_values("Popularity", ascending=False).head(top_n)
    print("Recommendations:\n", result, "\n")
    return result
```

```python
recommend_destination(df, best_time="Nov-Feb", dest_type="Historical", top_n=5)
```

# Main Function

```python
def main():
    # Step 1: Load + clean
    df = load_data()              # yahan koi quotes ki zarurat nahi
    df = clean_data(df)

    # Step 2: Analysis
    avg_dest = average_popularity_by_destination(df)
    avg_state = average_popularity_by_state(df)
    avg_type = average_popularity_by_type(df)

    # Step 3: Plots
    plot_avg_popularity_by_destination(avg_dest, top_n=5)
    plot_avg_popularity_by_state(avg_state)
    plot_avg_popularity_by_type(avg_type)

    # Step 4: Recommendations
    print("Example 1: Best time = 'Nov-Feb', Type = 'Historical'")
    recommend_destination(df, best_time="Nov-Feb", dest_type="Historical", top_n=5)

    print("Example 2: Best time = 'Apr-Jun', Type = 'Adventure'")
    recommend_destination(df, best_time="Apr-Jun", dest_type="Adventure", top_n=5)

if __name__ == "__main__":
    main()
```
